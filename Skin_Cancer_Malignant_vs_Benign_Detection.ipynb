##ImageDataGenerator

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#for training images
train_datagen = ImageDataGenerator(rotation_range=20, 
                               width_shift_range=0.10, 
                               height_shift_range=0.10,
                               shear_range=0.1, 
                               zoom_range=0.1, 
                               horizontal_flip=True, 
                               fill_mode='nearest',
                               rescale = 1./255 
                              )

#for testing images
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
plt.imshow(imread(benign))
```

```python
plt.imshow(train_datagen.random_transform(imread(benign)))
```

## Creating the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Dropout, Flatten, Dense, Conv2D, MaxPooling2D, BatchNormalization
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
optimizer = Adam(lr = 0.001)
```

```python
model = Sequential()

model.add(Conv2D(filters = 16, kernel_size = (3, 3), input_shape = image_shape, padding = 'Same', activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.3))

model.add(Conv2D(filters = 32, kernel_size = (3, 3), padding = 'Same', activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.35))

model.add(Conv2D(filters = 64, kernel_size = (3, 3), padding = 'Same',  activation = 'relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Dropout(0.4))


model.add(Flatten())


model.add(Dense(128, activation = 'relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(64, activation = 'relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy',
              optimizer = optimizer ,
              metrics = ['accuracy'])
```

```python
model.summary()
```

## Early Stopping

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = "val_accuracy", min_delta= 0.00005, patience= 7, verbose= 1, restore_best_weights= True)
```

```python
reduce_lr = ReduceLROnPlateau(monitor = 'val_accuracy', factor = 0.5, patience = 5, min_lr = 1e-7,verbose=1)
```

## Training the Model

```python
batch_size = 32
```

```python
image_shape[:2]
```

```python
train_image_gen = train_datagen.flow_from_directory(train_path,
                                               target_size=image_shape[:2],
                                                color_mode='rgb',
                                               batch_size=batch_size,
                                               class_mode='binary')
```

```python
test_image_gen = test_datagen.flow_from_directory(test_path,
                                               target_size=image_shape[:2],
                                               color_mode='rgb',
                                               batch_size=batch_size,
                                               class_mode='binary',shuffle=False)
```

```python
train_image_gen.class_indices
```

```python
model.fit_generator(train_image_gen,epochs=100,
                              validation_data=test_image_gen,
                             callbacks=[early_stop, reduce_lr])
```

## **For Hyperparameter Tuning using Keras-Tuner**

```python
from tensorflow import keras
from kerastuner import RandomSearch
from kerastuner.engine.hyperparameters import HyperParameters

def build_model(hp):
  model = keras.Sequential([
                           keras.layers.Conv2D(
                                        filters = hp.Int('conv_1_filter',    
                                        min_value=16,
                                        max_value=64,
                                        step=32), 
                                        kernel_size = hp.Int('conv_1_filter',
                                        min_value=3,
                                        max_value=5,
                                        step=1),activation='relu',padding = 'same',input_shape=image_shape),
                                   


                            keras.layers.Conv2D(
                                        filters = hp.Int('conv_2_filter',    
                                        min_value=32,
                                        max_value=128,
                                        step=32), 
                                        kernel_size = hp.Int('conv_2_filter',
                                        min_value=3,
                                        max_value=5,
                                        step=1),activation='relu',padding = 'same'),
                            

                            keras.layers.Conv2D(
                                        filters = hp.Int('conv_3_filter',    
                                        min_value=32,
                                        max_value=128,
                                        step=32), 
                                        kernel_size = hp.Int('conv_3_filter',
                                        min_value=3,
                                        max_value=5,
                                        step=1),activation='relu',padding = 'same'),
                                  


                            keras.layers.Flatten(),
                                   
                                   
                            keras.layers.Dense(
                                        units = hp.Int('dense_1_units',
                                        min_value=32,
                                        max_value=128,
                                        step=32), activation='relu'),
                                  
                            keras.layers.Dense(1, activation='sigmoid'),
  ])

  model.compile(
        optimizer=keras.optimizers.Adam(
            hp.Choice('learning_rate',
                      values=[1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy'])
  
  return model


tuner = RandomSearch(
                       build_model,
                       objective= 'val_accuracy',
                       max_trials= 5,
                       executions_per_trial=3,
                       directory = 'skin_cancer_detection',
                       project_name = 'skin_cancer')
  

tuner.search(train_image_gen,
               epochs = 5,
               validation_data = test_image_gen)
```
